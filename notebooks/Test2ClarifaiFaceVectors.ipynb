```python
image_url = 'https://s3-us-west-2.amazonaws.com/jchang-testbucket2-oregon/Faces/4014080-skina-heada1-eye4-eyebrow5-mouthb1-facehair4-nose2-hair2!white-none.jpg'
result3 = model.predict_by_url(url=image_url)
display(HTML(('<img src="%s" width=300 height=300/>' % (image_url))))
regions3 = result3['outputs'][0]['data'].get('regions')
if regions3:
    print('Number of faces detected:', len(result3['outputs'][0]['data']['regions']))
else:
    print('No face detection.')
vector3 = regions3[0]['data']['embeddings'][0]['vector']
print('Vector length:', len(vector3))
```

```python
'''
Try plotting the vector visually to discern anything. Nothing obvious.
'''

from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook

p = figure(title="test", x_axis_label='index', y_axis_label='vector', width=1500, height=300)
p.line(range(len(vector2)), vector2, line_width=1)
p.line(range(len(vector3)), vector3, line_width=1, line_color='red')
output_notebook()
show(p)
```
