```python
"""
TEST1
Test against AWS Rekognition.
Pre-created bucket=jchang-testbucket2-oregon and images uploaded into Faces folder.
Call index_faces on each image in Faces folder.
"""

import boto3
import pprint

REGION='us-west-2'
BUCKET='jchang-testbucket2-oregon'
COLLECTION='FaceTest'
FOLDER='FacesBigTest/'

def index_faces(bucket, key, image_id=None, collection_id=COLLECTION, attributes=(), region=REGION):
	rekognition = boto3.client("rekognition", region)
	response = rekognition.index_faces(
		Image={
			"S3Object": {
				"Bucket": bucket,
				"Name": key,
			}
		},
		CollectionId=collection_id,
#		ExternalImageId=image_id,
	    DetectionAttributes=attributes,
	)
	return response['FaceRecords']

    
s3 = boto3.resource('s3')
mybucket = s3.Bucket(BUCKET)
response = None
match_count = 0
total_count = 0
matched = {}
unmatched = {}
responses = {}
assert FOLDER, "FOLDER must be assigned with trailing '/'."
for index, object in enumerate(mybucket.objects.filter(Prefix=FOLDER)):
    print(str(index),'key:',object.key)
    if object.key.endswith('/'):
        continue # skip folders
    response = index_faces(BUCKET, object.key)
    if len(response) == 0:
        print("NO FACES")
        unmatched[object.key] = object
    else:
        pprint.pprint(response[0]['Face'])
        match_count += 1
        matched[object.key] = object
        responses[object.key] = response
    total_count += 1

print('Total:', total_count, 'Identified faces:', match_count)

    
```

```python
"""
Try grouping matched and unmatched faces. See if there is any discernible pattern.
Matches listed by sorted reverse confidence.

"""
from IPython.core.display import display, HTML
URL_PREFIX='https://s3-us-west-2.amazonaws.com/jchang-testbucket2-oregon/'
print('MATCHES:', match_count)
sorted_responses = sorted(responses.items(), key=lambda val: val[1][0]['Face']['Confidence'], reverse=True)
htmlstr = ''
for key,srvalues in sorted_responses:
    htmlstr += '<figure style="display:inline-block"><img src="{0}{1}" width=100 height=100 />\n'.format(URL_PREFIX,key)
    htmlstr += '<figcaption>{:.1f}</figcaption></figure>\n'.format(srvalues[0]['Face']['Confidence'])
display(HTML(htmlstr))

print('NO FACE MATCHES:', total_count-match_count)
htmlstr = ''
for key in unmatched:
    htmlstr += '<figure style="display:inline-block"><img src="{0}{1}" width=100 height=100 /></figure>\n'.format(URL_PREFIX,key)
display(HTML(htmlstr))
```
